```python
import pandas as pd
import matplotlib.pyplot as plt
import jupyternotify

df = pd.read_csv('AMiner-Coauthor.txt', delimiter = "\t",header=None,
                 names=['First author', 'Second author', 'Number of collaborations'])

df['First author'] = df['First author'].str[1:]
df = df.sort_values(by='Number of collaborations', ascending=False)
df = df.reset_index(drop=True)
df = df.apply(pd.to_numeric)

pd.set_option('float_format', '{:f}'.format)
```

```python
import networkx as nx
G = nx.Graph()

for i in range(len(df)):
    G.add_edge(int(df['First author'][i]), int(df['Second author'][i]), weight=int(df['Number of collaborations'][i]))

```

```python

print(f'Number of nodes: {nx.number_of_nodes(G)}')
print(f'Number of edges: {nx.number_of_edges(G)}')
print(f'Density: {nx.density(G)}')
print(f'Average Clustering: {nx.average_clustering(G)}')
```
